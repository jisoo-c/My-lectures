파이토치(pytorch)

```python
from tensorflow import keras
import numpy as np
```

```python
x = np.array([0, 1, 2, 3, 4], dtype='float64') #학습(train) 데이터. dtype은 데이터타입, float은 '실수'를 뜻함.
y = x * 2 + 1 #정답(label) 데이터. 정답을 지정했으므로 지도학습.
print(y)
```

```python
model = keras.models.Sequential() # 모델 레이어 만들기(--> 모델을 어떻게 할거냐)
model.add(keras.layers.Dense(1, input_shape=(1,)))
#'input_shape은 모델의 입력 데이터의 형태나 차원을 지정(튜플 형태). 여기서는 1*1 행렬. --> 종속/독립변수 1개씩 --> 1차원 배열
model.compile('SGD', 'mse') #모델 컴파일(--> 학습을 어떻게 할거냐)
#SGD는 '확률적 경사 하강법(Stochastic Gradient Descent)'으로 머신/딥러닝 최적화에 자주 쓰임. (참고: "미니배치")
#mse는 mean squared error
model.fit(x, y, epochs=200) #epoch=사이클 수.
```

```python
#학습을 시켰으니 이제 예측을 시켜보자
model.predict([6, 28, 51, 89, 876])
```

# 여러 층 쌓기

```python
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#텍스트
texts = ["이 영화 정말 좋아요!", "별로에요.", "스토리가 흥미롭지 않아요.", "재미있어요."]
# labels = [1, 0, 0, 1]  # 긍정(1), 부정(0)
labels = np.array([1, 0, 0, 1], dtype=np.int32)  # 레이블을 NumPy 배열로 변환
```

```python
#전처리
max_words = 1000  #최대 단어 수
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
padded_sequences = pad_sequences(sequences, maxlen=10, padding='post', truncating='post')
```

```python
#모델정의
model = keras.Sequential([
    keras.layers.Embedding(input_dim=max_words, output_dim=300, input_length=10), #입력층. max_words 길이로 200차원에 임베딩함. 시퀀스 길이는 10.
    keras.layers.Flatten(), #1차원벡터로 펼쳐주는 층
    keras.layers.Dense(32, activation='relu'), #은닉층
    keras.layers.Dense(32, activation='relu'), #은닉층2
     keras.layers.Dense(32, activation='relu'), #은닉층3
    keras.layers.Dense(1, activation='sigmoid') #출력층(확률값 출력)
])
```

```python
#컴파일
model.compile(optimizer='adam', #최적화
              loss='binary_crossentropy',  #이진분류이므로 binary_crossentropy 사용
              metrics=['accuracy']) #정확도를 기준으로 평가(정확도는 "맞는 것을 맞게 예측한 것")
```

```python
#모델학습
model.fit(padded_sequences, labels, epochs=100, batch_size=3)
```

```python
#예측할 텍스트
new_texts = ["이 영화 정말 재미있고 최고에요"]

#텍스트를 토큰화하고 패딩
new_sequences = tokenizer.texts_to_sequences(new_texts)
new_padded_sequences = pad_sequences(new_sequences, maxlen=10, padding='post', truncating='post')

#예측
predictions = model.predict(new_padded_sequences)

#예측결과
for i, text in enumerate(new_texts):
    sentiment = "긍정" if predictions[i] > 0.5 else "부정"
    print(f"텍스트: {text}")
    print(f"감정 예측: {sentiment} (확률: {predictions[i][0]:.2f})")
```
